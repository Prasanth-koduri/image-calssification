```
import os
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```
x = []
y = []
data1 = '/content/drive/MyDrive/Photo/mini_pro_2/benign_2'
data2 = '/content/drive/MyDrive/Photo/mini_pro_2/malignant_2'
```

```
for i in os.listdir(data1):
  img = imread(os.path.join(data1,i))
  img_re = resize(img,(150,150,3))
  img_flat = img_re.flatten()
  x.append(img_flat)
  y.append(0)
```

```
for i in os.listdir(data2):
  img = imread(os.path.join(data2,i))
  img_re = resize(img,(150,150,3))
  img_flat = img_re.flatten()
  x.append(img_flat)
  y.append(1)
```

```
x = np.array(x)
y = np.array(y)
print(x,y)
```

```
x.shape
```

```
y.shape
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=100)
```

```
from sklearn.model_selection import GridSearchCV
from sklearn import svm
p = [
              {'C':[1,10,100],'kernel':['linear','rbf']}
]
svc = svm.SVC(probability=True)
clf = GridSearchCV(svc,p)
clf.fit(x_train,y_train)
```

```
y_pred = clf.predict(x_test)
y_pred
```

```
y_test
```

```
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_pred,y_test)
```

```
confusion_matrix(y_pred,y_test)
```

```
import pickle
pickle.dump(clf,open('mini_model_2.p','wb'))
```

```
model = pickle.load(open('mini_model_2.p','rb'))
```
